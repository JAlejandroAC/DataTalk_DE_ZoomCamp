```python
import pandas as pd
```

```python
url = 'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet'
```

```python
df = pd.read_parquet(
    url
)
```

```python
df.head()
```

```python
pd.set_option('display.max_columns', None)

df.head()
```

```python
q1 = df[(df['lpep_pickup_datetime'] >= '2025-11-01') & 
                  (df['lpep_pickup_datetime'] <= '2025-12-01') & 
                  (df['trip_distance'] <= 1)]

print(f"Number of trips: {len(q1)}")
```

```python
q2 = df[(df['trip_distance'] <= 100)]

print(f"Longest trip distance and day: {q2['trip_distance'].max()} , {q2.loc[q2['trip_distance'].idxmax()]['lpep_pickup_datetime']}")
```

```python
url2 = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv'
```

```python
df2 = pd.read_csv(
    url2
)

print(df2.head())
```

```python
df_joined = df.merge(df2, left_on='PULocationID', right_on='LocationID', how='left')
df_joined.head()
```

```python
# Filter for pickups on Nov 18, 2025
df_nov18 = df_joined[(df_joined['lpep_pickup_datetime'] >= '2025-11-18') &
                     (df_joined['lpep_pickup_datetime'] < '2025-11-19')]

# Find the zone with the highest total amount
zone_max_total = df_nov18.groupby('Zone')['total_amount'].sum().idxmax()
max_total = df_nov18.groupby('Zone')['total_amount'].sum().max()
print(f"Zone with the most total amount on Nov 18, 2025: {zone_max_total} (${max_total:.2f})")
```

```python
df_eharlem = df_joined[(df_joined['Zone'] == 'East Harlem North')]

zone_max_total = df_nov18.groupby('DOLocationID')['tip_amount'].sum().idxmax()

df_q4 = df2[df2['LocationID'] == zone_max_total]

print(f"Zone with the highest total tip amount on Nov 18, 2025: {df_q4['Zone'].values[0]}")
```
